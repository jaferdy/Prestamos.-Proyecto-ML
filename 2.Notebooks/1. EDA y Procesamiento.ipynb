<span style="color: yellow;">Calculamos la proporciÃ³n entre el "Funded Amount" y el "Interest Rate" para capturar la relaciÃ³n entre el monto del prÃ©stamo y el interÃ©s aplicado. Esto podrÃ­a reflejar la carga financiera asociada al prÃ©stamo.</span>

```python
df["Funded Amount/Interest Rate"] = df["Funded Amount"]/df["Interest Rate"]
```

<span style="color: yellow;">Calculamos la proporciÃ³n entre el "Funded Amount" y el "Debit to Income" para obtener una medida de la relaciÃ³n entre el monto del prÃ©stamo y los ingresos del prestatario. Esto podrÃ­a indicar la capacidad de pago del prestatario en funciÃ³n de sus ingresos.</span>

```python
df["Funded Amount / Debit to Income"] = df["Funded Amount"] / df["Debit to Income"]
```

<span style="color: yellow;">Calculamos la proporciÃ³n entre "Inquires - six months" y "Open Account". Esto podrÃ­a indicar la actividad financiera del prestatario en los Ãºltimos seis meses.</span>


```python
df["Inquires - six months / Open Account"] = df["Inquires - six months"] / df["Open Account"]
```

<span style="color: yellow;">Calcula la proporciÃ³n entre el "Revolving Balance" y el "Total Revolving Credit Limit" para obtener una medida de la utilizaciÃ³n del lÃ­mite de crÃ©dito disponible del prestatario. Esto podrÃ­a indicar la proporciÃ³n de deuda pendiente en relaciÃ³n con el lÃ­mite de crÃ©dito.</span>

```python
df["Revolving Balance / Total Revolving Credit Limit"] = df["Revolving Balance"] / df["Total Revolving Credit Limit"]
```

<span style="color: yellow;">Calculamos la proporciÃ³n de "Delinquency - two years" y "Total Accounts" (cuentas en morosidad respecto al total de cuentas del prestatario). Esto podrÃ­a indicar el historial de pagos y la estabilidad financiera del prestatario.</span>

```python
df["Delinquency - two years / Total Accounts"] = df["Delinquency - two years"] / df["Total Accounts"]
```

<span style="color: yellow;">Creamos una nueva caracterÃ­stica que represente la cantidad total de cuentas pÃºblicas del prestatario, sumando las columnas "Public Record" y "Collection 12 months Medical". Esto podrÃ­a ser indicativo del historial crediticio y la gestiÃ³n financiera del prestatario.</span>

```python
df["cuentas pÃºblicas prestatario"] = df["Public Record"] + df["Collection 12 months Medical"]
```

<span style="color: yellow;">Calculamos la proporciÃ³n entre el monto total recibido ("Total Received Interest" + "Total Received Late Fee") y el monto del prÃ©stamo ("Funded Amount"). Esto podrÃ­a reflejar la eficacia de los pagos realizados por el prestatario en relaciÃ³n con el monto originalmente prestado.</span>

```python
df["Eficacia pagos respecto financiaciÃ³n inicial"] = (df["Total Received Interest"] + df["Total Received Late Fee"]) / df["Funded Amount"]
```
