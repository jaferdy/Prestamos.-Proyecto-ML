<span style="color: yellow;">Calculamos la proporción entre el "Funded Amount" y el "Interest Rate" para capturar la relación entre el monto del préstamo y el interés aplicado. Esto podría reflejar la carga financiera asociada al préstamo.</span>

```python
df["Funded Amount/Interest Rate"] = df["Funded Amount"]/df["Interest Rate"]
```

<span style="color: yellow;">Calculamos la proporción entre el "Funded Amount" y el "Debit to Income" para obtener una medida de la relación entre el monto del préstamo y los ingresos del prestatario. Esto podría indicar la capacidad de pago del prestatario en función de sus ingresos.</span>

```python
df["Funded Amount / Debit to Income"] = df["Funded Amount"] / df["Debit to Income"]
```

<span style="color: yellow;">Calculamos la proporción entre "Inquires - six months" y "Open Account". Esto podría indicar la actividad financiera del prestatario en los últimos seis meses.</span>


```python
df["Inquires - six months / Open Account"] = df["Inquires - six months"] / df["Open Account"]
```

<span style="color: yellow;">Calcula la proporción entre el "Revolving Balance" y el "Total Revolving Credit Limit" para obtener una medida de la utilización del límite de crédito disponible del prestatario. Esto podría indicar la proporción de deuda pendiente en relación con el límite de crédito.</span>

```python
df["Revolving Balance / Total Revolving Credit Limit"] = df["Revolving Balance"] / df["Total Revolving Credit Limit"]
```

<span style="color: yellow;">Calculamos la proporción de "Delinquency - two years" y "Total Accounts" (cuentas en morosidad respecto al total de cuentas del prestatario). Esto podría indicar el historial de pagos y la estabilidad financiera del prestatario.</span>

```python
df["Delinquency - two years / Total Accounts"] = df["Delinquency - two years"] / df["Total Accounts"]
```

<span style="color: yellow;">Creamos una nueva característica que represente la cantidad total de cuentas públicas del prestatario, sumando las columnas "Public Record" y "Collection 12 months Medical". Esto podría ser indicativo del historial crediticio y la gestión financiera del prestatario.</span>

```python
df["cuentas públicas prestatario"] = df["Public Record"] + df["Collection 12 months Medical"]
```

<span style="color: yellow;">Calculamos la proporción entre el monto total recibido ("Total Received Interest" + "Total Received Late Fee") y el monto del préstamo ("Funded Amount"). Esto podría reflejar la eficacia de los pagos realizados por el prestatario en relación con el monto originalmente prestado.</span>

```python
df["Eficacia pagos respecto financiación inicial"] = (df["Total Received Interest"] + df["Total Received Late Fee"]) / df["Funded Amount"]
```

<span style="color: yellow;">En primer lugar:</span>

<span style="color: yellow;">Ordenamos Grade y SubGrade de 0 a *n*, siendo *n* el mejor valor. El mejor valor será el de A en Grade y A1 en SubGrade. Las agencias de calificación clasifican como *A* a los prestatarios con menor indice de riesgo percibido. En SubGrade ocurre lo mismo.</span>

<span style="color: yellow;">Ordenamos Verification Status de forma en que el valor mayor será el de Source Verified, significa que la información ha sido verificada directamente de la fuente, lo que implica un mayor grado de confianza en su autenticidad y precisión, seguido por "Verified", que indica que la información ha sido verificada, pero no se especifica si fue obtenida directamente de la fuente o a través de otros medios, y por último "Not Verified", que significa que la información no ha sido verificada o que no se ha confirmado su exactitud.</span>

<span style="color: yellow;">Home Ownership es ordenado de mayor a menor grado de probabilidad de morosidad. </span>

<span style="color: yellow;">"Initial List Status":</span>

<span style="color: yellow;">"W" (Waiting): Indica que el préstamo se encuentra en espera y no ha sido activado ni procesado completamente. Esto significa que el préstamo ha sido presentado pero aún no está disponible para los inversionistas o prestamistas para financiamiento. Puede estar pendiente de verificación, revisión o aprobación antes de pasar a la siguiente etapa. Lo catalogaremos con un 0.</span> 

<span style="color: yellow;">"F" (Forwarded): Indica que el préstamo ha sido enviado o transmitido a los inversionistas o prestamistas para su revisión y posible financiamiento. En este estado, el préstamo está disponible para que los inversionistas consideren invertir en él y puede estar en proceso de recibir ofertas de financiamiento. Lo catalogaremos con un 1.</span>

```python
sub_grade_ordered = {"G5":0,"G4":1,"G3":2,"G2":3,"G1":4,
                     "F5":8,"F4":9,"F3":10,"F2":11,"F1":12,
                     "E5":16,"E4":17,"E3":18,"E2":19,"E1":20,
                     "D5":25,"D4":26,"D3":27,"D2":28,"D1":29,
                     "C5":35,"C4":36,"C3":37,"C2":38,"C1":39,
                     "B5":45,"B4":46,"B3":47,"B2":48,"B1":49,
                     "A5":55,"A4":56,"A3":57,"A2":58,"A1":59}
#Asignamos valor a dichos valores de menor a mayor. En economía, el mayor valor crediticio se corresponde a la letra A.
df["Sub Grade_ordered"] = df["Sub Grade"].map(sub_grade_ordered) # Mapeamos valores anteriores

grade_ordered = {"G":0,"F":2,"E":5,"D":8,"C":12,"B":16,"A":20} # Hacemos lo mismo que en sub grade por la misma razón.
df["Grade_ordered"] = df["Grade"].map(grade_ordered)

verification_status_ordered = {"Not Verified":0,
                               "Verified":2,
                               "Source Verified": 4}
df["verification_status_ordered"] = df['Verification Status'].map(verification_status_ordered)

home_ownership_ordered = {"RENT":1,
                          "MORTGAGE":2,
                          "OWN":5} 
df["Home Ownership ordered"] = df["Home Ownership"].map(home_ownership_ordered)

initial_list_status_ordered = df.groupby("Initial List Status")["Loan Status"].mean()
initial_list_status_ordered = dict(initial_list_status_ordered)
df["initial_list_status_ordered"] = df["Initial List Status"].map(initial_list_status_ordered)

application_type_ordered = df.groupby("Application Type")["Loan Status"].mean()
application_type_ordered = dict(application_type_ordered)
df["application_type_ordered"] = df["Application Type"].map(application_type_ordered)
```
